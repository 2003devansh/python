Dictionaries :- A dictionary is a collection which is unordered, chaneable and indexed, in python dictionaries are written with curly brackets, and they have key values pair

```python
dict = {"Car1" : "Audi" , "Car2" : "BMW", "Car3" : "Mercedes Benz"}
type(dict)
print(dict)
```

```python
# We can access the item values based on the keys
dict['Car1']
```

```python
 # we can even loop through-out the dictionaries keys
for x in dict: 
     print(x)
```

```python
 # we can even loop through-out the dictionaries values 
for x in dict.values() : 
     print(x)
```

```python
# we can also check both keys and values 
for x in dict.items():
    print(x)
```

```python
# Adding item in dictionaries
dict["Car4"]="Audi S-class"
print(dict)
```

```python
# nested dictionary 
car1_model =  {"mercedes" : 1960}
car2_model = {"Audi" : 1970}
car3_model = {"Ambassador" : 1980} 

dict = {"Car1" : car1_model , "Car2" : car2_model ,"Car3" : car3_model}
print(dict)
```
